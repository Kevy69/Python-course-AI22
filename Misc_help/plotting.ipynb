# Plotting

- x = [x1, x2, ... ,xn]
- y = [y1, y2, ... ,yn]

```py
plt.plot(x,y)
```

````python
import matplotlib.pyplot as plt 

pichu_height = [21, 21, 20, 15, 10]
pichu_width = [28, 31, 15, 15, 10]
pikachu_height = [31, 51, 40, 35, 20]
pikachu_width = [25, 29, 40, 45, 30]


plt.plot(pichu_width, pichu_height, '*')
plt.plot(pikachu_width, pikachu_height, 'o')
plt.legend(["Pichu", "Pikachu"])
````
